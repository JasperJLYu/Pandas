```python
# 导包
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 基础画图
plt.plot([1, 2, 4, 9, 5, 3])
plt.show()
```

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

```python
x = np.linspace(-2, 2, 500)
y = x ** 2
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
# plt默认画的是点之间的直线
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "r--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()

plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

```python
# 一幅图当中多个子图
x = np.linspace(-1.4, 1.4, 30)
plt.Figure((20, 20))
plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.title("figure_1")

plt.subplot(2, 2, 2)
plt.plot(x, x ** 2)
plt.title("figure_2")

plt.subplot(2, 2, 3)
plt.plot(x, x ** 3)
plt.title("figure_3")

plt.subplot(2, 2, 4)
plt.plot(2, 2, 4)
plt.title("figure_4")

plt.show()
```
